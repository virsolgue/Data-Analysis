```python
def analizar_val_ausentes(data_set):
  #Se verifica el número de valores ausentes para cada variable.
  #Esto se hace reconociendo lo valores nulos de cada celda.
  print ("Cantidad de valores ausentes por variable:")
  print("-------------------------------------------------------------------------------------")
  for columna in range (0,CANTIDAD_COLUMNAS):
    counter=0
    for fila in range (1,len(data_set)):
      if data_set[fila][columna] == '':
        counter+=1
    print(columna," - " , data_set[0][columna], ": ", counter, sep=' ', end='\n')
  print()

analizar_val_ausentes(data_set)

def reemplazar_nulos_por_nan(data_set):
  #Se reemplaza los valores nulos de la tabla por "nan".
  #Esto se hace reconociendo lo valores nulos de cada celda.
  for columna in range (0,CANTIDAD_COLUMNAS):
    for fila in range (1,len(data_set)):
      if data_set[fila][columna] == '':
        data_set[fila][columna] = np.nan
  return data_set

data_set_sin_nulos = reemplazar_nulos_por_nan(data_set)

```
